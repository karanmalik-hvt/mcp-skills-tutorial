```python
def validate_skill_frontmatter(frontmatter: dict, folder_name: str = None) -> dict:
    errors = []
    warnings = []
    
    # Validate 'name'
    name = frontmatter.get("name")
    if not name:
        errors.append("Missing required field: 'name'")
    else:
        if len(name) > 64:
            errors.append(f"Name too long: {len(name)} chars (max 64)")
        if not re.match(r'^[a-z0-9]([a-z0-9-]*[a-z0-9])?$', name):
            errors.append(f"Invalid name format: '{name}' â€” must be lowercase, hyphens, no leading hyphen")
        if '--' in name:
            errors.append(f"Name contains consecutive hyphens: '{name}'")
        if folder_name and name != folder_name:
            errors.append(f"Name '{name}' doesn't match folder '{folder_name}'")
    
    # Validate 'description'
    desc = frontmatter.get("description")
    if not desc:
        errors.append("Missing required field: 'description'")
    else:
        if len(desc) > 1024:
            errors.append(f"Description too long: {len(desc)} chars (max 1024)")
        lower_desc = desc.lower()
        if not any(kw in lower_desc for kw in ["when", "use for", "use when", "activate"]):
            warnings.append("Description doesn't explain WHEN to use this skill")
    
    # Validate optional fields
    compat = frontmatter.get("compatibility", "")
    if compat and len(compat) > 500:
        errors.append(f"Compatibility too long: {len(compat)} chars (max 500)")
    
    return {"valid": len(errors) == 0, "errors": errors, "warnings": warnings}
```

```python
validate_script = '''
#!/usr/bin/env python3
"""Validate a database migration plan."""

import json
import sys

def validate_plan(plan_path: str) -> bool:
    with open(plan_path) as f:
        plan = json.load(f)
    
    migrations = plan.get("migrations", [])
    if not migrations:
        print("ERROR: No migrations found in plan.")
        return False
    
    errors = []
    warnings = []
    required_fields = ["table", "change_type", "reversible"]
    
    for i, migration in enumerate(migrations):
        for field in required_fields:
            if field not in migration:
                errors.append(f"Migration {i+1}: missing required field '{field}'")
        if migration.get("reversible") is False:
            warnings.append(
                f"Migration {i+1}: {migration.get('change_type')} on "
                f"'{migration.get('table')}' is NOT reversible"
            )
    
    for w in warnings:
        print(f"WARNING: {w}")
    for e in errors:
        print(f"ERROR: {e}")
    
    if errors:
        print(f"\\nValidation FAILED: {len(errors)} error(s)")
        return False
    
    print(f"Validation PASSED: {len(migrations)} migration(s), {len(warnings)} warning(s)")
    return True

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python validate.py <plan.json>")
        sys.exit(1)
    success = validate_plan(sys.argv[1])
    sys.exit(0 if success else 1)
'''
```
